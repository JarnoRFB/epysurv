# eypsurv demo
This notes demonstrates the basic usage of `epysurv`

```python
import matplotlib.pyplot as plt
import pandas as pd
from epysurv import data as epidata
from epysurv.visualization.model_diagnostics import plot_prediction
plt.rc("figure", figsize=(16, 8))
```

Let's first get some data and plot it. We use the case counts of Salmonella Newport from Germany between 2004 and 2013. The data is already split into training and test set. We see that there are no outbreaks in the training set, but sone around the end of 2011.

```python
train, test = epidata.salmonella()

fig, ax = plt.subplots()
train.add_suffix("_train").plot(drawstyle="steps-post", ax=ax)
test.add_suffix("_test").plot(drawstyle="steps-post", ax=ax)
```

Nex we import some classical models: `EarsC1` and `FarringtonFlexible`. The API works really similiar to `scikit-learn`. Finally, we plot the predictions.

```python
from epysurv.models.timepoint import EarsC1, FarringtonFlexible
```

```python
model = EarsC1()
model.fit(train)
pred = model.predict(test)
plot_prediction(train, test, pred);
```

```python
model = FarringtonFlexible()
model.fit(train)
pred = model.predict(test)
plot_prediction(train, test, pred);
```
